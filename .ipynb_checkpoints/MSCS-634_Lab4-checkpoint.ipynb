```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# Load the Diabetes dataset
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = pd.DataFrame(diabetes.target, columns=['progression'])

# Display dataset information
print("Features (X) head:")
print(X.head())
print("\nTarget (y) head:")
print(y.head())
print("\nFeatures (X) info:")
X.info()
print("\nTarget (y) info:")
y.info()
print("\nFeatures (X) describe:")
print(X.describe())
print("\nTarget (y) describe:")
print(y.describe())

# Missing values in the dataset.
print("\nMissing values in X:")
print(X.isnull().sum())
print("\nMissing values in y:")
print(y.isnull().sum())
```

```python
# Here no explicit cleaning is needed for this dataset.
# In case of missing values some common strategies include:
# X.fillna(X.mean(), inplace=True) for numerical features
# X.dropna(inplace=True) for dropping rows with missing values
```

```python
# Step 2: Linear Regression.
# Selecting one feature for simple linear regression('bmi')
X_simple = X[['bmi']]
y_simple = y

# Spliting the data into training and testing sets
X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(
    X_simple, y_simple, test_size=0.2, random_state=42
)

# Linear Regression model
lin_reg_simple = LinearRegression()
lin_reg_simple.fit(X_train_simple, y_train_simple)

# Make predictions
y_pred_simple = lin_reg_simple.predict(X_test_simple)

# Evaluating the model
mae_simple = mean_absolute_error(y_test_simple, y_pred_simple)
mse_simple = mean_squared_error(y_test_simple, y_pred_simple)
rmse_simple = np.sqrt(mse_simple)
r2_simple = r2_score(y_test_simple, y_pred_simple)

print("\n Simple Linear Regression Metrics (using 'bmi')")
print(f"MAE: {mae_simple:.4f}")
print(f"MSE: {mse_simple:.4f}")
print(f"RMSE: {rmse_simple:.4f}")
print(f"R-squared: {r2_simple:.4f}")

# Visualizing the the model's prediction's agains the t actual data
plt.figure(figsize=(10, 6))
plt.scatter(X_test_simple, y_test_simple, color='red', label='Actual Data')
plt.plot(X_test_simple, y_pred_simple, color='blue', linewidth=2, label='Linear Regression Prediction')
plt.title('Simple Linear Regression: BMI vs. Disease Progression')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 3: Multiple Regression.
# Using all the features for multiple regression
X_multi = X
y_multi = y

# Spliting the data into training and testing sets
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X_multi, y_multi, test_size=0.2, random_state=42
)

# Multiple Regression model
multi_reg = LinearRegression()
multi_reg.fit(X_train_multi, y_train_multi)

# Making predictions
y_pred_multi = multi_reg.predict(X_test_multi)

mae_multi = mean_absolute_error(y_test_multi, y_pred_multi)
mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
rmse_multi = np.sqrt(mse_multi)
r2_multi = r2_score(y_test_multi, y_pred_multi)

print("\n Multiple Linear Regression Metrics (using all features)")
print(f"MAE: {mae_multi:.4f}")
print(f"MSE: {mse_multi:.4f}")
print(f"RMSE: {rmse_multi:.4f}")
print(f"R-squared: {r2_multi:.4f}")

# Visualize predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test_multi, y_pred_multi, alpha=0.7)
plt.plot([y_test_multi.min(), y_test_multi.max()], [y_test_multi.min(), y_test_multi.max()], 'r--', lw=2)
plt.title('Multiple Linear Regression: Actual vs. Predicted Values')
plt.xlabel('Actual Disease Progression')
plt.ylabel('Predicted Disease Progression')
plt.grid(True)
plt.show()
```

```python
# Step 4: Polynomial 
# Using all features for polynomial regression
X_poly = X
y_poly = y

# Spliting the data into training and testing data,
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(
    X_poly, y_poly, test_size=0.2, random_state=42
)

# Experimenting with different polynomial degrees
degrees = [1, 2, 3, 4] # Here degree 1 is equivalent to Linear Regression

print("\n Polynomial Regression Metrics")
results_poly = {}

for degree in degrees:
    print(f"\n Degree {degree}")
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("Polynomial_features", polynomial_features),
                         ("Linear_regression", linear_regression)])
    pipeline.fit(X_train_poly, y_train_poly)
    y_pred_poly = pipeline.predict(X_test_poly)

    mae_poly = mean_absolute_error(y_test_poly, y_pred_poly)
    mse_poly = mean_squared_error(y_test_poly, y_pred_poly)
    rmse_poly = np.sqrt(mse_poly)
    r2_poly = r2_score(y_test_poly, y_pred_poly)
    results_poly[degree] = {'MAE': mae_poly, 'MSE': mse_poly, 'RMSE': rmse_poly, 'R2': r2_poly}

    print(f"MAE: {mae_poly:.4f}")
    print(f"MSE: {mse_poly:.4f}")
    print(f"RMSE: {rmse_poly:.4f}")
    print(f"R-squared: {r2_poly:.4f}")

    # Visualizing the actual vs. predicted for each of the degree 
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test_poly, y_pred_poly, alpha=0.7)
    plt.plot([y_test_poly.min(), y_test_poly.max()], [y_test_poly.min(), y_test_poly.max()], 'r--', lw=2)
    plt.title(f'Polynomial Regression (Degree {degree}): Actual vs. Predicted')
    plt.xlabel('Actual Disease Progression')
    plt.ylabel('Predicted Disease Progression')
    plt.grid(True)
    plt.show()

# Discussion of overfitting/underfitting:
degrees_list = list(results_poly.keys())
r2_scores_poly = [results_poly[d]['R2'] for d in degrees_list]

plt.figure(figsize=(8, 5))
plt.plot(degrees_list, r2_scores_poly, marker='o')
plt.title('R-squared Score vs. Polynomial Degree')
plt.xlabel('Polynomial Degree')
plt.ylabel('R-squared Score')
plt.xticks(degrees_list)
plt.grid(True)
plt.show()
```

```python
# Observation on Polynomial Degree and Overfitting/Underfitting
# As the polynomial degree increases, the model becomes more complex.
# A degree of 1 is equivalent to linear regression.
# Higher degrees (e.g., 3 or more) can lead to overfitting if the model captures noise in the training data.
# This might result in a high R-squared on the training set but poor performance 
# lower R-squared, higher errors on the unseen test set."
# Conversely, a very low degree (e.g., 1 if the true relationship is non-linear) might lead to underfitting.
```

```python
# Step 5: Regularization with Ridge and Lasso Regression
# Using all features for regularization models
X_reg = X
y_reg = y

# Spliting the data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

# Defining a range of alpha values to observe their influence
alphas = [0.1, 1, 10, 100]

print("\nRidge Regression Metrics")
ridge_results = {}
for alpha in alphas:
    print(f"\n--- Ridge with alpha = {alpha} ---")
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train_reg, y_train_reg)
    y_pred_ridge = ridge_model.predict(X_test_reg)

    mae_ridge = mean_absolute_error(y_test_reg, y_pred_ridge)
    mse_ridge = mean_squared_error(y_test_reg, y_pred_ridge)
    rmse_ridge = np.sqrt(mse_ridge)
    r2_ridge = r2_score(y_test_reg, y_pred_ridge)

    ridge_results[alpha] = {'MAE': mae_ridge, 'MSE': mse_ridge, 'RMSE': rmse_ridge, 'R2': r2_ridge, 'Coeffs': ridge_model.coef_}

    print(f"MAE: {mae_ridge:.4f}")
    print(f"MSE: {mse_ridge:.4f}")
    print(f"RMSE: {rmse_ridge:.4f}")
    print(f"R-squared: {r2_ridge:.4f}")
    # print(f"Coefficients: {ridge_model.coef_.flatten()}") # Uncomment to see coefficient values

print("\n Lasso Regression Metrics")
lasso_results = {}
for alpha in alphas:
    print(f"\n--- Lasso with alpha = {alpha} ---")
    lasso_model = Lasso(alpha=alpha, max_iter=10000) # Increase max_iter for convergence
    lasso_model.fit(X_train_reg, y_train_reg)
    y_pred_lasso = lasso_model.predict(X_test_reg)

    mae_lasso = mean_absolute_error(y_test_reg, y_pred_lasso)
    mse_lasso = mean_squared_error(y_test_reg, y_pred_lasso)
    rmse_lasso = np.sqrt(mse_lasso)
    r2_lasso = r2_score(y_test_reg, y_pred_lasso)

    lasso_results[alpha] = {'MAE': mae_lasso, 'MSE': mse_lasso, 'RMSE': rmse_lasso, 'R2': r2_lasso, 'Coeffs': lasso_model.coef_}

    print(f"MAE: {mae_lasso:.4f}")
    print(f"MSE: {mse_lasso:.4f}")
    print(f"RMSE: {rmse_lasso:.4f}")
    print(f"R-squared: {r2_lasso:.4f}")
    print(f"Coefficients: {lasso_model.coef_.flatten()}")

# Visualizing how regularization parameters influence model behavior (R-squared vs. alpha)
plt.figure(figsize=(12, 6))
plt.plot(list(ridge_results.keys()), [r['R2'] for r in ridge_results.values()], marker='o', label='Ridge R-squared')
plt.plot(list(lasso_results.keys()), [r['R2'] for r in lasso_results.values()], marker='x', label='Lasso R-squared')
plt.xscale('log') # Alpha values often span orders of magnitude
plt.title('R-squared vs. Alpha for Ridge and Lasso Regression')
plt.xlabel('Alpha (Log Scale)')
plt.ylabel('R-squared')
plt.legend()
plt.grid(True)
plt.show()

# Visualizing the coefficient magnitudes for a specific alpha to highlight Lasso sparsity
plt.figure(figsize=(14, 7))
plt.bar(diabetes.feature_names, ridge_results[1]['Coeffs'].flatten(), label='Ridge Coefficients (Alpha=1)')
plt.bar(diabetes.feature_names, lasso_results[1]['Coeffs'], label='Lasso Coefficients (Alpha=1)', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.title('Coefficients for Ridge and Lasso Regression (Alpha=1)')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


print("\nExplanation of Regularization Parameters (Alpha Values):**")
print("Alpha is the regularization strength. Higher alpha value increases the penalty on the magnitude of the coefficients.")

print("\nRidge Regression")
print("Adds a penalty equal to the square of the magnitude of coefficients to the loss function. It shrinks coefficients towards zero but does not set them exactly to zero.")
print("This helps in reducing multicollinearity and preventing overfitting by making the model less sensitive to individual data points.")
print("As alpha increases, coefficients get smaller, and the model becomes simpler (less prone to overfitting).")

print("Lasso Regression")
print("Adds a penalty equal to the absolute value of the magnitude of coefficients to the loss function. It can shrink some coefficients exactly to zero, effectively performing feature selection.")
print("This is useful when you suspect that only a subset of features are truly relevant.")
print("As alpha increases, more coefficients are forced to zero, leading to a sparser model.")

print("\nDifferences in Performance:")
print("Ridge tends to keep all features but shrinks their impact, while Lasso can eliminate features by setting their coefficients to zero.")
print("The choice between Ridge and Lasso often depends on whether feature selection is desired or if all features are believed to be relevant but need their impact scaled down.")
print("In cases with many irrelevant features, Lasso  outperform Ridge by simplifying the model.")
print("If all features are somewhat relevant, Ridge might be more stable.")
```

```python

#Model Comparison Summary
# Storing results in a dictionary for easy access
comparison_results = {
    'Simple Linear Regression (BMI)': {'MAE': mae_simple, 'MSE': mse_simple, 'RMSE': rmse_simple, 'R2': r2_simple},
    'Multiple Linear Regression (All Features)': {'MAE': mae_multi, 'MSE': mse_multi, 'RMSE': rmse_multi, 'R2': r2_multi},
}

# Add Polynomial Regression results (e.g., for optimal or a representative degree)
# Here, let's pick the degree with the highest R2 or a specific degree for comparison
best_poly_degree = max(results_poly, key=lambda k: results_poly[k]['R2'])
comparison_results[f'Polynomial Regression (Degree {best_poly_degree})'] = results_poly[best_poly_degree]

# Add Ridge and Lasso Regression results (e.g., for optimal alpha or a representative alpha)
# For simplicity, let's choose alpha=1 or the alpha that yielded the highest R2
best_ridge_alpha = max(ridge_results, key=lambda k: ridge_results[k]['R2'])
best_lasso_alpha = max(lasso_results, key=lambda k: lasso_results[k]['R2'])
comparison_results[f'Ridge Regression (Alpha={best_ridge_alpha})'] = ridge_results[best_ridge_alpha]
comparison_results[f'Lasso Regression (Alpha={best_lasso_alpha})'] = lasso_results[best_lasso_alpha]

print("Model Comparison")
for model_name, metrics in comparison_results.items():
    print(f"\n{model_name}:")
    for metric_name, value in metrics.items():
        if isinstance(value, np.ndarray): # Handle cases where coefficients might be in the dictionary
            print(f"  {metric_name}: {np.round(value, 4)}")
        else:
            print(f"  {metric_name}: {value:.4f}")
```

```python
# How well each model performed:
# - Simple Linear Regression: As expected, it performed the worst, with the lowest R-squared (0.2334) and highest error metrics (MAE: 52.26, MSE: 4061.83). This clearly shows that a single feature (BMI) is insufficient to model the complexity of diabetes progression.
# - Multiple Linear Regression: Showed a significant improvement over simple linear regression, nearly doubling the R-squared to 0.4526 and considerably reducing error metrics (MAE: 42.79, MSE: 2900.19). This indicates that considering multiple physiological factors is crucial for better prediction.
# - Polynomial Regression (Degree 1): As it's equivalent to Multiple Linear Regression for degree 1, its performance (R2: 0.4526) was identical. If higher degrees were explored, we would expect to see potential for capturing non-linearities, but also a higher risk of overfitting if the degree is too high and not supported by the underlying data patterns.
# - Ridge Regression (Alpha=0.1): Achieved a slightly better R-squared (0.4609) and slightly lower errors (MAE: 42.99, MSE: 2856.49) compared to standard Multiple Linear Regression. This suggests that even a small amount of L2 regularization helped in producing a marginally more robust model, potentially by reducing the impact of multicollinearity among features.
# - Lasso Regression (Alpha=0.1): Demonstrated the best overall performance among all models, with the highest R-squared (0.4719) and lowest error metrics (MAE: 42.85, MSE: 2798.19). Notably, Lasso also performed implicit feature selection by driving some coefficients to exactly zero, indicating that certain features might be less relevant for prediction in this context.

# Which models handled overfitting or improved performance:
# - Regularization (Ridge and Lasso): These models were instrumental in improving performance and handling potential overfitting. By introducing a penalty term, they constrained the model's coefficients. Lasso, in particular, not only improved R-squared but also simplified the model by setting less impactful feature coefficients to zero, which is a strong defense against overfitting by reducing model complexity. Ridge, by shrinking coefficients, also contributed to a more generalized model.
# - Polynomial Regression (careful degree selection): While the degree 1 case showed no improvement, higher degrees of polynomial regression are designed to capture non-linear relationships. If such non-linearities exist, choosing an appropriate (not excessively high) degree could improve performance without overfitting, which would typically involve cross-validation for optimal degree selection. The current results don't show an improvement with degree 1, but the principle holds.

# Insights gained about the Diabetes dataset:
# - The progression of diabetes is clearly a complex, multifactorial condition. A single predictor like BMI is highly insufficient, and incorporating multiple physiological measurements is essential for building a predictive model.
# - The dataset's characteristics seem to benefit from regularization. The improvements seen with Ridge and especially Lasso suggest that either there's some multicollinearity among features, or some features have limited predictive power, which Lasso effectively identifies and prunes.
# - Lasso's ability to zero out coefficients (e.g., for 'age', 's4', and 's7' based on the typical feature order) provides valuable insights into which physiological measurements might be less critical for predicting disease progression in this specific linear model context. This could be a significant finding for further medical research.
# - The dataset appears relatively well-behaved for demonstrating various regression techniques, as all models produced sensible results, and regularization provided tangible benefits.
```
